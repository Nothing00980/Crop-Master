```python
dict = {"Jan":1,"Feb":2,"Mar":3,"Apr":4,"May":5,"Jun":6,"Jul":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}
```

```python
month_column = []
```

```python
# month = [i.month for i in data['Date']]

for i in data2['Date']:
    str = i
    str2 = str.split('-')
    month_column.append(dict[str2[1]])
```

```python
data2['month_column'] = month_column
```

```python
data2
```

```python
season_names= []
```

```python
for k in data2['month_column']:
    if k==1 or k==2:
        season_names.append("winter")
    elif k==3 or k==4:
        season_names.append("spring")
    elif k==5 or k==6:
        season_names.append("summer")
    elif k==7 or k==8:
        season_names.append("monsoon")
    elif k==9 or k==10:
        season_names.append("autumn")
    elif k==11 or k==12:
        season_names.append("pre winter")
    
```

```python
data2['season_names'] = season_names
```

```python
data2
```

```python
dayofweek = []
```

```python
for l in data2['Date']:
    d = pd.Timestamp(l)
    day = d.dayofweek
    dayofweek.append(day)
```

```python
data2['day'] = dayofweek
```

```python
data2
```

```python
year = []

```

```python
for k in data2['Date']:
    str1 = k
    str2 = str1.split('-')
    year.append(int(str2[2]))
```

```python
data2['year'] = year
```

```python
q1 = np.percentile(data2['Modalp'],25,interpolation="midpoint")
q2 = np.percentile(data2['Modalp'],75,interpolation="midpoint")
iqr = q2-q1
```

```python
upper = np.where(data2['Modalp'] >= (q2 + 1.5*iqr))
lower = np.where(data2['Modalp'] <= (q1 - 1.5*iqr))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(features2,label,test_size=0.1,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
regs = RandomForestRegressor(max_depth=1000,random_state=0)
regs.fit(xtrain,ytrain)
```

```python
yprediction = regs.predict(xtest)
from sklearn.metrics import r2_score
```

```python
r2_score(ytest,yprediction)
```
